```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("digits.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
pixels = df.drop("number_label", axis = 1)
```

```python
pixels.iloc[0]
```

```python
np.array(pixels.iloc[0])
```

```python
data = np.array(pixels.iloc[0]).reshape((8,8))
```

```python
pixel_plot, ax = plt.subplots()
plt.title("pixel_plot")
pixel_plot = plt.imshow(data,cmap="gray")
plt.colorbar(pixel_plot)
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        ax.text(j, i, str(data[i, j]), color='r', ha='center', va='center')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
```

```python
scaled_x = scalar.fit_transform(pixels)
```

```python
from sklearn.decomposition import PCA
```

```python
pca_model = PCA(n_components=2)
```

```python
pred = pca_model.fit_transform(scaled_x)
```

```python
np.sum(pca_model.explained_variance_ratio_)
```

```python
pca_model.components_
```

```python
df_comp = pd.DataFrame(pca_model.components_, index = ["PC1", "PC2"], columns=pixels.columns)
```

```python
df_comp
```

```python
pca_model.explained_variance_ratio_
```

```python
plt.figure(figsize=(20,8), dpi = 200)
plt.scatter(pred[:,0], pred[:,1], c = df["number_label"])
```

```python
pca_model = PCA(n_components=3)
pred = pca_model.fit_transform(scaled_x)
np.sum(pca_model.explained_variance_ratio_)
```

```python
pca_model.components_
```

```python
df_comp = pd.DataFrame(pca_model.components_, index = ["PC1", "PC2","PC3"], columns=pixels.columns)
```

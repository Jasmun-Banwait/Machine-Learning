```python
circles = pd.read_csv("cluster_circles.csv")
```

```python
circles.head()
```

```python
sns.scatterplot(data  = circles, x="X1", y = "X2")
```

```python
def display_categories(model,data):
    labels = model.fit_predict(data)
    sns.scatterplot(data=data, x = "X1", y = "X2", hue = labels, palette = "Set1")
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters= 2)
```

```python
display_categories(model,moons)
```

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(eps=0.2)
```

```python
display_categories(model, circles)
```

```python
out_per = []
num_out = []
for i in np.linspace(0.01,10,100):
    model = DBSCAN(eps=i)
    labels = model.fit_predict(two_blobs_out)
    num_out.append(np.sum(model.labels_ == -1))
    per_out = 100 * np.sum(model.labels_ == -1) / len(model.labels_)
    out_per.append(per_out)
```

```python
sns.lineplot(x = np.linspace(0.1,10,100), y = num_out)
```
